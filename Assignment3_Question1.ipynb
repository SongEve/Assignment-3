### 1. MaxPool


```python
maxpool2d=torch.nn.MaxPool2d(kernel_size=2, stride=1, padding=0,dilation=1, return_indices=False, ceil_mode=False)
torch_out=maxpool2d(Input)
torch_out
```

```python
def mymaxpool2d(Input, kernal_size, stride):
    batch,channel,height,width = Input.shape
    
    out_height = int((height-kernal_size)/stride)+1
    out_width = int ((width-kernal_size)/stride)+1
    print("out_shape", (out_height, out_width))
    out_pooling = torch.zeros((batch,channel, out_height, out_width))

    for img in range(batch):
        for map_num in range(channel):
            for h in range(out_height):
                for w in range(out_width):
                    out_pooling[img,map_num, h, w] = torch.max(Input[img,map_num, (h*stride):(h + kernal_size), (w*stride):(w + kernal_size)])
    return out_pooling
```

```python
my_maxpool2d=mymaxpool2d(Input,2,1)
my_maxpool2d
```

### 3. Conv2d_1

```python
Conv2d=torch.nn.Conv2d(in_channels=3, out_channels=6,kernel_size=3, stride=1, padding=0, dilation=1,
                groups=1,bias=True, padding_mode='zeros')
torch_out=Conv2d(Input)
torch_out
```

```python
conv_weight = Conv2d.weight.data
conv_bias = Conv2d.bias.data
conv_weight.shape
```

```python
def myconv2d(Input, conv_weight, conv_bias, stride):
    batch,channel,height,width = Input.shape
    out_channel,_,k_h,k_w=conv_weight.shape
    
    out_height = int((height-k_h)/stride)+1
    out_width = int ((width-k_w)/stride)+1
    out_pooling = torch.zeros(batch,out_channel, out_height, out_width)

    for img in range(batch):
        for k in range(out_channel):
            for h in range(out_height):
                for w in range(out_width):
                    out_pooling[img,k, h, w] = torch.sum(Input[img,:, (h*stride):(h*stride + k_h), (w*stride):(w*stride +k_w )]
                                                                          *conv_weight[k,:,:,:])+conv_bias[k]
    return out_pooling
my_conv2d = myconv2d(Input, conv_weight, conv_bias, 1)
my_conv2d
```

### 4. Conv2d_2 (stride=2,dilation=2)

```python
Conv2d_2=torch.nn.Conv2d(in_channels=3, out_channels=6,kernel_size=5, stride=2, padding=0, dilation=2, groups=1,
bias=True, padding_mode='zeros')
torch_out=Conv2d_2(Input)
torch_out
```

```python
conv_weight = Conv2d_2.weight.data
conv_bias = Conv2d_2.bias.data
```

```python
def myconv2d_2(x, in_channels,out_channel,stride,bias,dilation):
    batch,channel,height,width = x.shape 
    
    k_h=conv_weight.shape[2]
    k_w=conv_weight.shape[3]
    
    dilation_k_h=k_h*dilation-1
    dilation_k_w =k_w*dilation-1
    
    out_height = int((height - dilation_k_h) / stride) + 1
    out_width = int((width - dilation_k_w) / stride) + 1
    out_pooling = torch.zeros(batch, out_channel, out_height, out_width)

    kernal = torch.zeros(out_channel, channel, dilation_k_h, dilation_k_w)

    for oc in range(out_channel):
        for c in range(channel):
            for kh in range(k_h):
                for kw in range(k_w):
                    kernal[oc, c, kh * dilation, kw * dilation] = conv_weight[oc, c, kh, kw]
    
    for img in range(batch):
        for oc in range(out_channel):
            for c in range(channel):
                for oh in range(out_height):
                    for ow in range(out_width):
                          out_pooling[img, oc, oh, ow] += (x[img, c, (oh * stride):(oh*stride + dilation_k_h), 
                                                 (ow * stride):(ow * stride + dilation_k_w)] * kernal[oc][c]).sum()
    
    for i in range(out_channel):
        out_pooling[0][i] += conv_bias[i]

    return out_pooling

myconv2D = myconv2d_2(Input,3, 6,2,  conv_bias,  2)

print(myconv2D)
```

### 5. Transpose2d

```python
transpose2d=torch.nn.ConvTranspose2d(in_channels=3, out_channels=4,kernel_size=3, stride=1, padding=0, output_padding=0,
groups=1, bias=True, dilation=1, padding_mode='zeros')
torch_out=transpose2d(Input)
torch_out
```

```python
conv_weight=transpose2d.weight.data
conv_bias=transpose2d.bias.data
```

```python
def mytranspose2d(in_channel,out_channel,stride,bias,dilation, Input):

    batch,channel,height,width = Input.shape
    k_h=conv_weight.shape[2]
    k_w=conv_weight.shape[3]
    
    out_height = (height + k_h) - 1
    out_width = (width + k_w) - 1
    out_pooling = torch.zeros(batch, out_channel, out_height, out_width)
    
    for img in range(batch):
        for k in range(out_channel):
            for c in range(channel):
                for h in range(height):
                    for w in range(width):
                          out_pooling[img, k, h:h + k_h, w:w + k_w] += Input[img, c, h, w] * conv_weight[c][k]
    
    for i in range(out_channel):
        out_pooling[0][i] += conv_bias[i]

    return out_pooling

myconv2D = mytranspose2d(3, 4, 1,conv_bias, 1,Input)

print(myconv2D)
```

### 8.roi Pool 

```python
boxes = torch.Tensor([[0, 0, 0, 6.5, 6.5]])
torchvision.ops.roi_pool(Input,boxes,output_size=(3,3))
```

```python

def my_roi_pool(Input, boxes, output_size):
    
    start_h = int(boxes[0][1])
    start_w = int(boxes[0][2])
    end_h = int(boxes[0][3]) 
    end_w = int(boxes[0][4])
    
    width = (end_w - start_w + 1)
    height = (end_h - start_h + 1)

    bin_w = width / output_size[0]
    bin_h = height / output_size[1]
    
    result = []
    for k in range(Input.shape[1]):
        temp = []
        for j in range(output_size[0]):
            sub_res = []
            for i in range(output_size[1]):
                sub_res.append(Input[..., int(0+bin_h*(j)):math.ceil(0+bin_h*(j+1)),
                                     int(0+bin_w*(i)):math.ceil(0+bin_w*(i+1))][0][k].max())
            temp.append(sub_res)
        result.append(temp)
    res=torch.Tensor(result)
    res = res.unsqueeze(0)
    return res
```

```python
my_roi_pool(Input,boxes, (3,3))
```

### 9.batch_norm

```python
channels = Input.shape[1]
mean = torch.Tensor([Input[0][i].mean() for i in range(Input.shape[1])])
var = torch.Tensor([Input[0][i].var() for i in range(Input.shape[1])])
```

```python
torch.nn.functional.batch_norm(Input, running_mean=mean, running_var=var,weight=None, bias=None, training=False, momentum=0.1,eps=1e-05)
```

```python
def batch_norm(Input,mean,var, momentum, eps):
    output = torch.zeros(Input.shape)
    for i in range(Input.shape[1]):
        output[0][i] = (Input[0][i] - mean[i])/(np.sqrt(var[i])-eps)
    return output
batch_norm(Input, mean, var,momentum=0.1, eps=1e-05)
```

### 10. Cross_entropy


```python
target = torch.randint(3, (1,32,32))
Input = torch.randn(1, 3, 32, 32)
target
```

```python
torch_out= torch.nn.functional.cross_entropy(Input, target, weight=None,
size_average=None, ignore_index=-100, reduce=None,reduction='mean') 
torch_out
```

```python
def softmax(x): 
    sof=torch.zeros(x.shape)
    for i in range (len(x)):
        loss=x[i].exp()/sum(x[i].exp())
        sof[i]=-torch.log(loss)
    return sof
        
def nl(input, target): 
    nloss = torch.zeros([len(input)])
    for i in range(len(input)):
        L=0
        for j in range(target[i].shape[1]):
            for k in range(target.shape[2]):
                L -= input[i,target[i,j,k],j,k]
        nloss[i]=- L/target.shape[1]/target.shape[2]
    return nloss
                
def cross_entropy(input,target):
    soft=softmax(input)
    loss=nl(soft,target)
    return loss

cross_entropy(Input,target)
```
